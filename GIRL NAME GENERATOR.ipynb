```python
import numpy as np
import re
```

```python
pre_femalename = open('female_name.txt', encoding='utf8').read()
femalename = pre_femalename.split()
femalename = list(map(str.lower, femalename))
femalename
```

```python
female_name_corpus = []
for word in femalename:
    split = list(word)
    for i in range(len(split)-1):
        female_name_corpus.append((split[i], split[i+1]))
female_name_corpus
```

```python
alphabet = []

for i, j in female_name_corpus:
    i = re.sub('[-=.#/?:$}\s\']', '', i)
    j = re.sub('[-=.#/?:$}\s\']', '', j)
#    j = re.sub('[-=.#/?:$}]', '', j)
    alphabet.append(i)
    alphabet.append(j)
    
alphabet # 중복값의 향연
alphabet_core = list(set(alphabet))
alphabet_core
```

```python
word_dict_fin = {}
for word_1, word_2 in female_name_corpus:
    if word_1 in word_dict_fin.keys():
        word_dict_fin[word_1].append(word_2)
    else:
        word_dict_fin[word_1] = [word_2]
        
word_dict_fin
```

```python
first_word = np.random.choice(alphabet_core)
first_word
```

```python
chain_fin = [first_word]
n_words_fin = 4
chain_fin
```

```python
for i in range(n_words_fin):
    if chain_fin[-1] in word_dict_fin: # KEY값이 존재하는지 확인한다.
        ransam_fin = np.random.choice(word_dict_fin[chain_fin[-1]])
        chain_fin.append(ransam_fin)
        
# if 'b' in word_dict_fin:
#    print("happy")
# else:
#    print("sad")
```
